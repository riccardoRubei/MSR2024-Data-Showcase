```python
sets = {'Nintendo': [], 'PlayStation': [], 'Xbox': [], 'PC': []}
for file in os.listdir('final_dataset'):
    if 'all_games' in file:
        df = pd.read_csv(os.path.join("final_dataset", file))
        console = file.split("_")[2].split(".")[0]
        sets[console] = set(df['id'].tolist())
```
