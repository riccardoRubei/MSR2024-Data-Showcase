```python
import pandas as pd
import os
import numpy as np
import csv
```

```python
#for each subfolder in IGDB_metadata_merged
for folder in os.listdir('IGDB_metadata_merged'):
    total = pd.DataFrame()
    if folder != '.DS_Store':
        #for each file in subfolder
        for file in os.listdir('IGDB_metadata_merged/'+folder):
            if file != '.DS_Store':
                #load file
                df = pd.read_csv('IGDB_metadata_merged/'+folder+'/'+file, index_col=0)
                #add to total
                total = pd.concat([total,df])

    with open(os.path.join("final_dataset",f"all_games_{folder}.csv"), 'w', newline='', encoding='utf-8') as file:
                writer = csv.writer(file, quoting=csv.QUOTE_MINIMAL)
                # Write header
                writer.writerow(total.columns)
                # Write data
                for _, row in total.iterrows():
                    writer.writerow(row)
```
