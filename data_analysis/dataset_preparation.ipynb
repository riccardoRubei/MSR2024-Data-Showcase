```python
import pandas as pd
import os
import numpy as np
import csv
```

```python
#for each subfolder in IGDB_metadata_merged
for folder in os.listdir('IGDB_metadata_merged'):
    total = pd.DataFrame()
    if folder != '.DS_Store':
        #for each file in subfolder
        for file in os.listdir('IGDB_metadata_merged/'+folder):
            if file != '.DS_Store':
                #load file
                df = pd.read_csv('IGDB_metadata_merged/'+folder+'/'+file)
                #add to total
                total = pd.concat([total,df])

    with open(os.path.join("final_dataset",f"all_games_{folder}.csv"), 'w', newline='', encoding='utf-8') as file:
                writer = csv.writer(file, quoting=csv.QUOTE_MINIMAL)
                # Write header
                writer.writerow(total.columns)
                # Write data
                for _, row in total.iterrows():
                    writer.writerow(row)
```

```python
#read all csv files in final_dataset
all_files = os.listdir('../final_dataset')
all_files = [x for x in all_files if x != '.DS_Store']
all_files = [x for x in all_files if 'screenshots' not in x]
for file in all_files:
    print(file)
    p = pd.read_csv(os.path.join("..","final_dataset",file))
    display(p.head())
```
